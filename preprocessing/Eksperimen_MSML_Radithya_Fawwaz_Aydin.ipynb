## Text Cleaning

```python
import re
import nltk
from nltk.corpus import stopwords

# Download stopwords bahasa inggris
nltk.download('stopwords')

# Load daftar stopwords ke dalam sebuah set
stop_words = set(stopwords.words('english'))
```

```python
# Fungsi pembersihan teks
def clean_text(text):
    # 1. Hapus tag HTML
    text = re.sub(r'<[^>]+>', '', text)
    text = re.sub(r'&[a-zA-Z0-9]+;', '', text)
    
    # 2. Ubah ke huruf kecil
    text = text.lower()
    
    # 3. Hapus tanda baca dan angka
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # 4. Hapus stopwords
    words = text.split()
    cleaned_words = [word for word in words if word not in stop_words]
    
    # Gabungkan kembali kata-katanya menjadi satu string
    return " ".join(cleaned_words)

# Terapkan fungsi ini ke kolom combined_text
print("Memulai pembersihan teks...")
df['cleaned_text'] = df['combined_text'].apply(clean_text)
print("Pembersihan teks selesai.")

# Tampilkan hasilnya untuk perbandingan
print("\nDataframe dengan 'cleaned_text': ")
display(df[['combined_text', 'cleaned_text']].head())
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Inisiasasi LabelEncoder
le = LabelEncoder()

# Fit dan transform kolom 'category'
df['category_encoded'] = le.fit_transform(df['category'])

# Tampilkan hasil untuk perbandingan
print("DataFrame dengan 'category_encoded': ")
display(df[['category', 'category_encoded']].head())
```

```python
# Gunakan 'le.classes_' untuk menyimpan nama kategori asli sesuai urutan encoding
mapping = dict(zip(le.classes_, le.transform(le.classes_)))

print("\nPemetaan Kategori ke Angka: ")

# Konversi nilai numpy.int64 menjadi nilai int standar
mapping_serializable = {k: int(v) for k, v in mapping.items()}

# Tampilkan pemetaan
import json
print(json.dumps(mapping_serializable, indent=2))
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

# 1. Tentukan fitur (X) dan target (y)
X = df['cleaned_text']
y = df['category_encoded']

# 2. Lakukan pemisahan data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42, stratify=y)

# 3. Verifikasi hasil pemisahan
print("Bentuk dari data yang sudah dibagi")
print(f"X_train: {X_train.shape}")
print(f"y_train: {y_train.shape}")
print(f"X_test: {X_test.shape}")
print(f"y_test: {y_test.shape}")
```

```python
# Cek distribusi di data training
print("\nDistribusi kategori di Training Set (y_train):")
# 'normalize=True' akan menampilkan persentase (%)
print(y_train.value_counts(normalize=True).sort_index())

# Cek distribusi di data testing
print("\nDistribusi kategori di Test Set (y_test):")
print(y_test.value_counts(normalize=True).sort_index())
```

## Vektorisasi Fitur (TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Inisialisasi TFidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# 2. Lakukan fit dan transform hanya pada X_train
print("Menerapkan TF-IDF ke X_train...")
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)

# 3. Lakukan lagi tapi hanya pada X_test
print("Menerapkan TF-IDF ke X_test...")
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# 4. Verifikasi hasilnya
print("\nBentuk dari data TF-IDF")
print(f"X_train_tfidf: {X_train_tfidf.shape}")
print(f"X_test_tfidf: {X_test_tfidf.shape}")
```
